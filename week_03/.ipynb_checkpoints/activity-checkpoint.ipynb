```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import exposure
from skimage.exposure import cumulative_distribution
```

```python
imgCv = cv2.imread("pic.jpg")
imgPlt = plt.imread("pic.jpg")
```

```python
im = imgCv
fps = 10

a = 1.0
b = 0

Frame = []

for i in range(30):
    gamma = 0.1+(0.1*i)
    im_test = np.zeros((538,1024,3), dtype='uint8')
    im_test[:,:] = (a*(((im[:,:]/255)**gamma)+b))*255
    Frame.append(im_test)

file_path = 'gamma_mr_van.mp4'
fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
writer = cv2.VideoWriter(file_path, fourcc, fps, (1024, 538))

for frame in Frame:
    writer.write(frame)
    
for frame in reversed(Frame):
    writer.write(frame)

writer.release()
print("Finished")
```

```python
img_eqa = cv2.cvtColor(imgPlt, cv2.COLOR_BGR2RGB)

img_eqa[:,:,0] = cv2.equalizeHist(imgPlt[:,:,0])
img_eqa[:,:,1] = cv2.equalizeHist(imgPlt[:,:,1])
img_eqa[:,:,2] = cv2.equalizeHist(imgPlt[:,:,2])

f, axarr = plt.subplots(2,2)

f.set_figheight(15)
f.set_figwidth(15)
axarr[0,0].imshow(imgPlt)

# color = ('b','g','r')
# for i,col in enumerate(color):
#     histr = cv2.calcHist([imgCv],[i],None,[256],[0,256])
#     plt.plot(histr,color = col)
#     plt.xlim([0,256])
# plt.show()

histr_b = cv2.calcHist([imgCv],[0],None,[256],[0,256])
histr_g = cv2.calcHist([imgCv],[1],None,[256],[0,256])
histr_r = cv2.calcHist([imgCv],[2],None,[256],[0,256])

axarr[0,1].plot(histr_b, color = 'b')
axarr[0,1].plot(histr_g, color = 'g')
axarr[0,1].plot(histr_r, color = 'r')


axarr[1,0].imshow(img_eqa)

histr_eqa_b = cv2.calcHist([img_eqa],[0],None,[256],[0,256])
histr_eqa_g = cv2.calcHist([img_eqa],[1],None,[256],[0,256])
histr_eqa_r = cv2.calcHist([img_eqa],[2],None,[256],[0,256])

axarr[1,1].plot(histr_eqa_b, color = 'b')
axarr[1,1].plot(histr_eqa_g, color = 'g')
axarr[1,1].plot(histr_eqa_r, color = 'r')
```
